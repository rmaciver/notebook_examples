# Plotting Rheomtry Data

A simple example of importing an excel file and plotting the data with Pandas

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from glob import glob

from matplotlib import rc
rc('text', usetex=True)
```

```python
# two ways to load files:
# 1. Exact name in relative path, read as excel file:
df = pd.read_excel('data/SW2_sifted_27112019_rwd.xlsx')

# 2. With a glob pattern, read in a loop:
# note that these are simply copies of the same file to demonstrate the method
files = glob('data/*.xlsx')
dfs = [pd.read_excel(file) for file in files]
```

```python
# the columns have been renamed to a format easier for indexing the dataframe
print(df.keys())

# keep these here for reference:
# labels = {'tau' τ in N∕m²	ɣ̇ in 1∕s	η in mPas	T in °C	t in s	t_seg in s
```

```python
# clean up data data: 
# B2 contains a strange symbol, so skip that row
df = df[1:]
```

```python
# can extract the variables first, then plot them
t = df.t
tau = df.tau
velar = df.velar
eta = df.eta

f, ax = plt.subplots(1,2, dpi=250, figsize=(7, 2.2))

ax[0].plot(t, tau, marker='None', markersize=4, color='r',  label='$\\tau$')
ax[0].set_xlabel('Time, $t$ [s]')
ax[0].set_ylabel('Shear velocity, $\\tau$ [Nm$^{-2}$]')
ax[0].legend()

# note that eta was scaled by 1000 and the units converted to Pas
ax[1].plot(t, eta/1000, marker='None', markersize=4, color='g')
ax[1].set_xlabel('Time, $t$ [s]')
ax[1].set_ylabel('Kinematic viscosity, $\eta$ [Pas]')


plt.tight_layout()

save = True
if save: 
    f.savefig('data/rheo_data.png')
```

```python
# plot several times on the same figure, with some fake data

# make a new figure
f, ax = plt.subplots(1,1, dpi=250, figsize=(3.5, 2.2))

# make a fake scaling of tau and plot in a loop
for scale in [1, 0.9, 0.8, 0.7, 0.6]:
    label = '%0.1f' % scale
    ax.plot(t, tau*scale, marker='None', label=label)
    ax.set_xlabel('Time, $t$ [s]')
    ax.set_ylabel('Shear velocity, $\\tau$ [Nm$^{-2}$]')
    ax.legend()

# plotting in a loop can also be applied to different files, or different tabs in the same files!
save = True
if save: 
    f.savefig('data/rheo_data_scaled.png')
```
